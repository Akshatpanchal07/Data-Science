Total Number of Gas and Fuel Cars

```python
sns.countplot(x='fuel-type', data=df, palette= 'mako')
plt.xlabel('Fuel Type')
plt.ylabel('Count')
plt.title('Distribution of Cars Based on Fuel Type')
plt.legend(title='Fuel Type', labels=['Gasoline', 'Diesel'])

plt.show()
```

Separate the Numerical data 

```python
df.info()
```

```python
a = ['symboling', 'normalized-losses', 'width', 'height',
       'engine-size', 'horsepower', 'city-mpg', 'highway-mpg', 'price']
num = df[a]
```

```python
num.head()
```

Find the Relation-ship between the different numerical colmuns

```python
cr = num.corr()
sns.heatmap(data = cr, cmap = 'mako', annot = True)
plt.figure(figsize=(20, 20))
plt.show()
```

```python
plt.figure(figsize=(6, 6))
sns.scatterplot(x='engine-size', y='horsepower', data=df , hue= 'fuel-type', palette = 'mako' )
plt.xlabel('Engine Size')
plt.ylabel('Horsepower')
plt.title('Engine Size vs. Horsepower and Fuel Type' )
plt.show()
```

Car price distribution along with Kernel Density Estimate

```python
sns.histplot(df['price'], kde=True, color='blue')
plt.xlabel('Price')
plt.ylabel('Count')
plt.title('Distribution of Car Prices')

plt.show()
```

Show the Mileage of car in city as well as on Highway

```python
plt.figure(figsize=(8,6))
sns.scatterplot(x = 'highway-mpg',y = 'city-mpg', data = num, palette = 'mako')
plt.xlabel('Mileage on Highway')
plt.ylabel('Mileage in City')
plt.title('City Mileage Vs Highway Mileage')
plt.show()
```
